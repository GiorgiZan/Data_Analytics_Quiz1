# გამოიყენეთ კონკრეტული სვეტის მნიშვნელობისთვის სტატისტიკური ფუნქციები (mean, standard deviation, median, min, max).

```python
mean_value = df['WorkExp'].mean()
std_deviation = df['WorkExp'].std()
median_value = df['WorkExp'].median()
min_value = df['WorkExp'].min()
max_value = df['WorkExp'].max()
```

```python
# გამომაქვს სამუშაო გამოცდილების მინიმალური, მაქსიმულარი, საშუალო და სხვა სტატისტიკური მნიშნვლებობები
print(f"საშუალო: {mean_value}")
print(f"სტანდარტული გადახრა: {std_deviation}")
print(f"მედიანი: {median_value}")
print(f"მინიმალური: {min_value}")
print(f"მაქსიმალური: {max_value}")
```

# Numpy და Matplotlib-ის ბიბლიოთეკების გამოყენებით ააგეთ 2 სხვადასხვა ტიპის გრაფიკი (მაგ. Bar დახაზოვანი დიაგრამები).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# RemoteWork შემთხვევების დათვლა და ცვლადში შენახვა
remote_work_counts = df['RemoteWork'].value_counts()

# ბარ ტიპის დიაგრამა
plt.figure(figsize=(10, 6))
remote_work_counts.plot(kind='bar', color='orange')
plt.title('Distribution of Remote Work Status')
plt.xlabel('Remote Work Status')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ხაზოვანი დიაგრამა
sopartfreq_counts = df['SOPartFreq'].value_counts()

plt.figure(figsize=(12, 6))
sopartfreq_counts.plot(kind='line', marker='o', color='green', linestyle='-', linewidth=2, markersize=8)
plt.title('Distribution of Stack Overflow Participation Frequency')
plt.xlabel('Frequency of Participation')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python

median_salary = df.groupby('RemoteWork')['ConvertedCompYearly'].median().sort_values()

plt.figure(figsize=(10,6))
median_salary.plot(kind='line', marker='o', color='b')
plt.title('Median Salary by Remote Work Status')
plt.xlabel('Remote Work Status')
plt.ylabel('Median Salary (Yearly)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
